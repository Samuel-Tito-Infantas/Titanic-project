## 2) Se você fosse sugerir a priorização de algum grupo/cluster para ser iniciado o trabalho, qual seria? Quais variávies você utilizou para identificar esse grupo?

## Fazendo um score básico de doenças a faixa de 34 a 38 anos tem o maior risco.
## o risco foi calculado utilizando a sumatoria das viariáveis de forma simples. 

```python
gb = df.groupby('Faixa_etaria').sum()
gb = gb.iloc[:,[0,3,4,5,6,7,8,9,10]]
gb = gb[['Diabetes','Hipertensao',	'Cancer',	'Asma',	'Osteomuscular_risco',	'Gestante',	'Puépera',	'Saúde Mental']]
gb['Score_total'] = gb[['Diabetes','Hipertensao',	'Cancer',	'Asma',	'Osteomuscular_risco',	'Gestante',	'Puépera',	'Saúde Mental']].sum(axis=1)
```

```python
gb.sort_values(by='Score_total', ascending=False).head(3)
```
